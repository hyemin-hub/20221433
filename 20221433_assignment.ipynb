```python
### ë¬¸ì œ 1: LeetCode 225. Implement Stack using Queues


#queue í´ë˜ìŠ¤ë¥¼ ì •ì˜
class Queue:
    def __init__(self):
        # ë¦¬ìŠ¤íŠ¸ë¡œ í êµ¬í˜„ (ë§¨ ì•ì—ì„œ êº¼ë‚´ê³ , ë’¤ì— ë„£ìŒ)
        self.items = []

    def isEmpty(self):
        # íê°€ ë¹„ì—ˆëŠ”ì§€ í™•ì¸
        return len(self.items) == 0

    def enqueue(self, item):
        # íì˜ ë’¤ì— ìš”ì†Œ ì¶”ê°€
        self.items.append(item)

    def dequeue(self):
        # íì˜ ì•ì—ì„œ ìš”ì†Œ ì œê±°í•˜ê³  ë°˜í™˜ (FIFO)
        if not self.isEmpty():
            return self.items.pop(0)
        else:
            return None

    def peek(self):
        # íì˜ ì•ì— ìˆëŠ” ìš”ì†Œ í™•ì¸
        if not self.isEmpty():
            return self.items[0]
        else:
            return None

    def size(self):
        # íì— ìˆëŠ” ìš”ì†Œ ê°œìˆ˜ ë°˜í™˜
        return len(self.items)



class MyStack:
    def __init__(self):
        # í 2ê°œ ì‚¬ìš©
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x):
        # q2ì— ë¨¼ì € ìƒˆ ê°’ ì¶”ê°€
        self.q2.enqueue(x)

        # q1ì— ìˆëŠ” ê°’ì„ ì „ë¶€ q2ë¡œ ì´ë™
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())

        # q1ê³¼ q2ë¥¼ êµí™˜
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        # ê°€ì¥ ì•ì— ìˆëŠ” ê°’ ì œê±°í•˜ê³  ë°˜í™˜
        return self.q1.dequeue()

    def top(self):
        # ë§¨ ì• ê°’ë§Œ í™•ì¸
        item = self.q1.dequeue()
        self.q2.enqueue(item)
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())
        self.q1, self.q2 = self.q2, self.q1
        return item

    def empty(self):
        return self.q1.isEmpty()



### ë‘ ê°œì˜ íë¡œ ìŠ¤íƒ êµ¬ì¡° êµ¬í˜„
### push, pop, top, empty ë“± ê¸°ëŠ¥ êµ¬í˜„
### ì‹œê°„ë³µì¡ë„
#- push: O(n)
#- pop: O(1)
#- top: O(1)
#- empty: O(1)
```

```python
### ë¬¸ì œ 2: LeetCode 232. Implement Queue using Stacks

# ìˆ˜ì—…ì—ì„œ ë°°ìš´ ë¦¬ìŠ¤íŠ¸ ìŠ¤íƒ ì§ì ‘ ì •ì˜

class ListStack:
    def __init__(self):
        # ìŠ¤íƒ ë°ì´í„°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸ ìƒì„±
        self.items = []

    def isEmpty(self):
        # ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ìˆëŠ”ì§€ í™•ì¸
        return len(self.items) == 0

    def push(self, item):
        # ë¦¬ìŠ¤íŠ¸ ëì— í•­ëª© ì¶”ê°€
        self.items.append(item)

    def pop(self):
        # ë§ˆì§€ë§‰ í•­ëª© ì œê±° ë° ë°˜í™˜
        if not self.isEmpty():
            return self.items.pop()
        else:
            return None  # ë¹„ì—ˆìœ¼ë©´ None ë°˜í™˜

    def peek(self):
        # ìŠ¤íƒì˜ top ìš”ì†Œ í™•ì¸
        if not self.isEmpty():
            return self.items[-1]
        else:
            return None

    def size(self):
        # ìŠ¤íƒì— ìˆëŠ” ìš”ì†Œ ê°œìˆ˜ ë°˜í™˜
        return len(self.items)


class MyQueue:
    def __init__(self):
        self.in_stack = ListStack()
        self.out_stack = ListStack()

    def push(self, x):
        # ìƒˆ ë°ì´í„°ë¥¼ in_stackì— ë„£ê¸°
        self.in_stack.push(x)

    def pop(self):
        # out_stackì´ ë¹„ì–´ ìˆìœ¼ë©´ in_stackì—ì„œ ë°ì´í„°ë¥¼ ì˜®ê¸°ê¸°
        if self.out_stack.isEmpty():
            while not self.in_stack.isEmpty():
                self.out_stack.push(self.in_stack.pop())
        return self.out_stack.pop()

    def peek(self):
        if self.out_stack.isEmpty():
            while not self.in_stack.isEmpty():
                self.out_stack.push(self.in_stack.pop())
        item = self.out_stack.pop()
        self.out_stack.push(item)
        return item

    def empty(self):
        return self.in_stack.isEmpty() and self.out_stack.isEmpty()



#### â± ì‹œê°„ë³µì¡ë„
#- push: O(1)
#- pop: **í‰ê· ì ìœ¼ë¡œ O(1)** (ì‹¤ì œ ë³µì‚¬ëŠ” í•„ìš”í•  ë•Œë§Œ í•¨)
#- peek: O(1)
#- empty: O(1)

#### ğŸ“Œ ì‚¬ìš©í•œ ìë£Œêµ¬ì¡°
#- Python ë¦¬ìŠ¤íŠ¸ 2ê°œë¥¼ ìŠ¤íƒì²˜ëŸ¼ ì‚¬ìš© (append/pop)
```

```python
## ì—°ìŠµë¬¸ì œ :: 8ë²ˆ ë¬¸ì œ


class Deque:
    def __init__(self):
        self.__queue = []

    def enqueueRear(self, x):  # ë’¤ì— ì‚½ì… (íì—ì„œì˜ enqueueì™€ ê°™ìŒ)
        self.__queue.append(x)

    def enqueueFront(self, x):  # ì•ì— ì‚½ì…
        self.__queue.insert(0, x)

    def dequeueFront(self):  # ì•ì—ì„œ ì‚­ì œ (íì—ì„œì˜ dequeueì™€ ê°™ìŒ)
        if self.isEmpty():
            print("Deque is empty.")
            return None
        return self.__queue.pop(0)

    def dequeueRear(self):  # ë’¤ì—ì„œ ì‚­ì œ
        if self.isEmpty():
            print("Deque is empty.")
            return None
        return self.__queue.pop()

    def front(self):
        if self.isEmpty():
            return None
        return self.__queue[0]

    def rear(self):
        if self.isEmpty():
            return None
        return self.__queue[-1]

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()

    def printDeque(self):
        print("Deque from front:", end=' ')
        for item in self.__queue:
            print(item, end=' ')
        print()


#ì˜ˆì‹œí…ŒìŠ¤íŠ¸


def main():
    d = Deque()
    d.enqueueRear("Mon")
    d.enqueueRear("Tue")
    d.enqueueFront("Sun")
    d.printDeque()         # Sun Mon Tue
    print(d.dequeueRear())  # Tue
    print(d.dequeueFront()) # Sun
    d.printDeque()         # Mon

main()



### 08. ListQueueë¥¼ Dequeë¡œ í™•ì¥í•œ í´ë˜ìŠ¤ ë§Œë“¤ê¸°

#DequeëŠ” íë³´ë‹¤ ê¸°ëŠ¥ì´ ë” ë§ì€ ìë£Œêµ¬ì¡°ë¡œ,
#ì•ê³¼ ë’¤ ì–‘ìª½ì—ì„œ ì‚½ì…ê³¼ ì‚­ì œê°€ ëª¨ë‘ ê°€ëŠ¥í•˜ë‹¤.
#ListQueue í´ë˜ìŠ¤ëŠ” í•œìª½(ë’¤ì—ì„œ ë„£ê³  ì•ì—ì„œ ëºŒ)ë§Œ ì§€ì›í–ˆì§€ë§Œ,
#Deque í´ë˜ìŠ¤ì—ì„œëŠ” ë‹¤ìŒê³¼ ê°™ì€ ë©”ì„œë“œë¥¼ ìƒˆë¡œ ë§Œë“¤ì—ˆë‹¤:

#- `enqueueFront(x)` : ì•ì—ì„œ ì‚½ì… (ë¦¬ìŠ¤íŠ¸ì˜ ë§¨ ì•ì— ë„£ê¸°)
#- `dequeueRear()` : ë’¤ì—ì„œ ì‚­ì œ (ë¦¬ìŠ¤íŠ¸ì˜ ë§¨ ë’¤ì—ì„œ pop)
#- ê¸°ì¡´ì˜ `enqueue()` â†’ `enqueueRear()`,
#  `dequeue()` â†’ `dequeueFront()`ë¡œ ì´ë¦„ë§Œ ë°”ê¿ˆ

#ë¦¬ìŠ¤íŠ¸ë¥¼ ì‚¬ìš©í–ˆê¸° ë•Œë¬¸ì— ì‚½ì…/ì‚­ì œëŠ” ê°„ë‹¨í•˜ê²Œ ì²˜ë¦¬í•  ìˆ˜ ìˆë‹¤.
```
