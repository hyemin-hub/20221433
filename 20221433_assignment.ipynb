```python
### 문제 1: LeetCode 225. Implement Stack using Queues


#queue 클래스를 정의
class Queue:
    def __init__(self):
        # 리스트로 큐 구현 (맨 앞에서 꺼내고, 뒤에 넣음)
        self.items = []

    def isEmpty(self):
        # 큐가 비었는지 확인
        return len(self.items) == 0

    def enqueue(self, item):
        # 큐의 뒤에 요소 추가
        self.items.append(item)

    def dequeue(self):
        # 큐의 앞에서 요소 제거하고 반환 (FIFO)
        if not self.isEmpty():
            return self.items.pop(0)
        else:
            return None

    def peek(self):
        # 큐의 앞에 있는 요소 확인
        if not self.isEmpty():
            return self.items[0]
        else:
            return None

    def size(self):
        # 큐에 있는 요소 개수 반환
        return len(self.items)



class MyStack:
    def __init__(self):
        # 큐 2개 사용
        self.q1 = Queue()
        self.q2 = Queue()

    def push(self, x):
        # q2에 먼저 새 값 추가
        self.q2.enqueue(x)

        # q1에 있는 값을 전부 q2로 이동
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())

        # q1과 q2를 교환
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        # 가장 앞에 있는 값 제거하고 반환
        return self.q1.dequeue()

    def top(self):
        # 맨 앞 값만 확인
        item = self.q1.dequeue()
        self.q2.enqueue(item)
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())
        self.q1, self.q2 = self.q2, self.q1
        return item

    def empty(self):
        return self.q1.isEmpty()



### 두 개의 큐로 스택 구조 구현
### push, pop, top, empty 등 기능 구현
### 시간복잡도
#- push: O(n)
#- pop: O(1)
#- top: O(1)
#- empty: O(1)
```

```python
### 문제 2: LeetCode 232. Implement Queue using Stacks

# 수업에서 배운 리스트 스택 직접 정의

class ListStack:
    def __init__(self):
        # 스택 데이터를 저장할 리스트 생성
        self.items = []

    def isEmpty(self):
        # 리스트가 비어있는지 확인
        return len(self.items) == 0

    def push(self, item):
        # 리스트 끝에 항목 추가
        self.items.append(item)

    def pop(self):
        # 마지막 항목 제거 및 반환
        if not self.isEmpty():
            return self.items.pop()
        else:
            return None  # 비었으면 None 반환

    def peek(self):
        # 스택의 top 요소 확인
        if not self.isEmpty():
            return self.items[-1]
        else:
            return None

    def size(self):
        # 스택에 있는 요소 개수 반환
        return len(self.items)


class MyQueue:
    def __init__(self):
        self.in_stack = ListStack()
        self.out_stack = ListStack()

    def push(self, x):
        # 새 데이터를 in_stack에 넣기
        self.in_stack.push(x)

    def pop(self):
        # out_stack이 비어 있으면 in_stack에서 데이터를 옮기기
        if self.out_stack.isEmpty():
            while not self.in_stack.isEmpty():
                self.out_stack.push(self.in_stack.pop())
        return self.out_stack.pop()

    def peek(self):
        if self.out_stack.isEmpty():
            while not self.in_stack.isEmpty():
                self.out_stack.push(self.in_stack.pop())
        item = self.out_stack.pop()
        self.out_stack.push(item)
        return item

    def empty(self):
        return self.in_stack.isEmpty() and self.out_stack.isEmpty()



#### ⏱ 시간복잡도
#- push: O(1)
#- pop: **평균적으로 O(1)** (실제 복사는 필요할 때만 함)
#- peek: O(1)
#- empty: O(1)

#### 📌 사용한 자료구조
#- Python 리스트 2개를 스택처럼 사용 (append/pop)
```

```python
### 연습문제 3번:
# a 안에 있는 요소들을 하나씩 b에 enqueue 해주는 방식

#LinkedQueue클래스


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedQueue:
    def __init__(self):
        self.front = None
        self.rear = None

    def isEmpty(self):
        return self.front is None

    def enqueue(self, x):
        new_node = Node(x)
        if self.isEmpty():
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.isEmpty():
            raise Exception("Queue is empty")
        x = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return x

    def peek(self):
        if self.isEmpty():
            return None
        return self.front.data

    def dequeueAll(self):
        self.front = None
        self.rear = None

    def printQueue(self):  # 확인용
        current = self.front
        while current:
            print(current.data, end=' ')
            current = current.next
        print()

def copyLinkedQueue(a):
  # 새로운 빈 큐 b를 만든다
    b = LinkedQueue()
    # a의 front부터 시작해서 끝까지 따라간다
    current = a.front
    while current:
      # 현재 노드의 데이터만 b에 enqueue 한다 (노드는 새로 만들어짐)
        b.enqueue(current.data)  # 데이터만 복사
        current = current.next  # 다음 노드로 이동
    return b


a = LinkedQueue()
a.enqueue("apple")
a.enqueue("banana")
a.enqueue("cherry")

b = copyLinkedQueue(a)

# 각 큐 출력
a.printQueue()  # apple banana cherry
b.printQueue()  # apple banana cherry

print(a is b)  # False → 서로 다른 객체

```

```python
#연습문제:: 4번

class ListQueue:
    def __init__(self):
        self.__queue = []

    def enqueue(self, x):
        self.__queue.append(x)

    def dequeue(self):
        return self.__queue.pop(0)  # 맨 앞에서 삭제

    def front(self):
        if self.isEmpty():
            return None
        return self.__queue[0]

    def isEmpty(self):
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()
class StackUsingQueues:
    def __init__(self):
        self.q1 = ListQueue()
        self.q2 = ListQueue()

    def push(self, x):
        # 1. 새 데이터를 q2에 먼저 넣는다
        self.q2.enqueue(x)

        # 2. q1의 모든 원소를 q2로 옮긴다
        while not self.q1.isEmpty():
            self.q2.enqueue(self.q1.dequeue())

        # 3. q1, q2를 교체 (q1이 항상 메인 큐)
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        # 항상 q1에서 pop
        if self.q1.isEmpty():
            print("Stack is empty.")
            return None
        return self.q1.dequeue()

    def top(self):
        return self.q1.front()

    def isEmpty(self):
        return self.q1.isEmpty()

    def clear(self):
        self.q1.dequeueAll()


def main():
    s = StackUsingQueues()
    s.push("Mon")
    s.push("Tue")
    s.push("Wed")
    print("Top:", s.top())      # Wed
    print("Pop:", s.pop())      # Wed
    print("Pop:", s.pop())      # Tue
    s.push("Fri")
    print("Pop:", s.pop())      # Fri
    print("Pop:", s.pop())      # Mon
    print("Empty?", s.isEmpty())  # True

main()
```

```python
##연습문제 5번
class ListStack:
    def __init__(self):
        self.__stack = []

    def push(self, x):
        self.__stack.append(x)

    def pop(self):
        return self.__stack.pop()

    def top(self):
        if self.isEmpty():
            return None
        return self.__stack[-1]

    def isEmpty(self):
        return len(self.__stack) == 0

    def popAll(self):
        self.__stack.clear()


class QueueUsingTwoStacks:
    def __init__(self):
        self.s1 = ListStack()  # 입력 전용 스택
        self.s2 = ListStack()  # 출력을 위한 임시 스택

    def enqueue(self, x):
        # push()는 s1에 그대로 넣기만 하면 됨 (간단함)
        self.s1.push(x)

    def dequeue(self):
        # s2가 비었으면 s1의 모든 데이터를 s2로 옮김
        if self.s2.isEmpty():
            while not self.s1.isEmpty():
                self.s2.push(self.s1.pop())

        # 이제 s2에서 pop()하면 가장 먼저 들어온 값이 나옴
        if self.s2.isEmpty():
            print("Queue is empty.")
            return None
        return self.s2.pop()

    def isEmpty(self):
        return self.s1.isEmpty() and self.s2.isEmpty()

#왜 “비효율적”인가?
#enqueue()는 빠르지만
#dequeue() 할 때마다 스택 전체를 뒤집어야 해서 느림


def main():
    q = QueueUsingTwoStacks()
    q.enqueue("Mon")
    q.enqueue("Tue")
    q.enqueue("Wed")
    print(q.dequeue())  # Mon
    print(q.dequeue())  # Tue
    q.enqueue("Thu")
    print(q.dequeue())  # Wed
    print(q.dequeue())  # Thu
    print(q.dequeue())  # Queue is empty.

main()
```

```python
## 연습문제 06번

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedQueue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, x):
        new_node = Node(x)
        if self.isEmpty():
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.isEmpty():
            raise Exception("Queue is empty")
        x = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return x


### 06. Deque를 단일 연결리스트(LinkedQueue)로 구현할 때 시간복잡도

#이 장에서 배운 큐는 연결리스트로 만든 구조이고,
#앞(front)과 뒤(rear)를 포인터로 가리키고 있다.
#이 구조에서 Deque처럼 앞과 뒤 양쪽에서 삽입/삭제하는 기능을 만들면
#연산마다 속도가 조금씩 다르다.

#| 연산              | 설명                                        | 시간복잡도 |
#|-------------------|---------------------------------------------|------------|
#| enqueueRear(x)    | 뒤에 넣기 (원래 큐랑 같음)                  | O(1)       |
#| enqueueFront(x)   | 앞에 넣기 (새 노드를 front로 바꾸면 됨)      | O(1)       |
##| dequeueFront()    | 앞에서 빼기 (원래 큐랑 같음)                | O(1)       |
#| dequeueRear()     | 뒤에서 빼기 (rear 앞 노드를 찾으려면 순회 필요) | O(n)       |

### 쉽게 요약하자면:

#- **앞에서 넣고 빼는 건 다 빠름** (한 번에 가능)
#- **뒤에서 넣는 것도 빠름**
#- **근데 뒤에서 빼는 건 느림**
#  → 왜냐하면, **rear 앞에 있는 노드를 못 바로 찾고**,
 #   맨 앞부터 쭉 따라가야 하기 때문임

### 결론:
#- enqueueRear / enqueueFront / dequeueFront = O(1)
#- dequeueRear = O(n)
```

```python
##연습문제 7번

#핵심구조
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedListBasic:
    def __init__(self):
        self.head = None

    def insert(self, x):  # 앞에 삽입
        new_node = Node(x)
        new_node.next = self.head
        self.head = new_node

    def delete(self):  # 앞에서 삭제
        if self.head is None:
            return None
        x = self.head.data
        self.head = self.head.next
        return x

### 07. 단일 연결 리스트(LinkedListBasic)로 Deque 구현 시 시간 복잡도

#LinkedListBasic 클래스는 head(앞부분)만 가지고 있는
#단순한 단일 연결 리스트 구조이다. rear는 없다.

#이 구조를 이용해서 Deque처럼 앞뒤 양쪽에서 삽입/삭제를 한다고 하면,
#각 연산의 시간 복잡도는 다음과 같다:

#| 연산            | 설명                                         | 시간복잡도 |
#|-----------------|----------------------------------------------|-------------|
#| enqueueFront(x) | head 앞에 넣는 거니까 바로 가능              | O(1)        |
#| dequeueFront()  | head에서 빼는 거니까 바로 가능               | O(1)        |
#| enqueueRear(x)  | 맨 끝에 넣으려면 끝까지 따라가야 함         | O(n)        |
#| dequeueRear()   | 맨 끝을 지우려면 끝까지 가고 그 전 노드도 필요 | O(n)        |

### 쉽게 정리하면:

#- 앞에서 넣고 빼는 건 빠름
#- 뒤에서 넣거나 빼려면 하나씩 따라가야 해서 느림

#결론적으로, 앞쪽은 O(1), 뒤쪽은 O(n)이다.
#
```

```python
## 연습문제 :: 8번 문제


class Deque:
    def __init__(self):
        self.__queue = []

    def enqueueRear(self, x):  # 뒤에 삽입 (큐에서의 enqueue와 같음)
        self.__queue.append(x)

    def enqueueFront(self, x):  # 앞에 삽입
        self.__queue.insert(0, x)

    def dequeueFront(self):  # 앞에서 삭제 (큐에서의 dequeue와 같음)
        if self.isEmpty():
            print("Deque is empty.")
            return None
        return self.__queue.pop(0)

    def dequeueRear(self):  # 뒤에서 삭제
        if self.isEmpty():
            print("Deque is empty.")
            return None
        return self.__queue.pop()

    def front(self):
        if self.isEmpty():
            return None
        return self.__queue[0]

    def rear(self):
        if self.isEmpty():
            return None
        return self.__queue[-1]

    def isEmpty(self) -> bool:
        return len(self.__queue) == 0

    def dequeueAll(self):
        self.__queue.clear()

    def printDeque(self):
        print("Deque from front:", end=' ')
        for item in self.__queue:
            print(item, end=' ')
        print()


#예시테스트


def main():
    d = Deque()
    d.enqueueRear("Mon")
    d.enqueueRear("Tue")
    d.enqueueFront("Sun")
    d.printDeque()         # Sun Mon Tue
    print(d.dequeueRear())  # Tue
    print(d.dequeueFront()) # Sun
    d.printDeque()         # Mon

main()



### 08. ListQueue를 Deque로 확장한 클래스 만들기

#Deque는 큐보다 기능이 더 많은 자료구조로,
#앞과 뒤 양쪽에서 삽입과 삭제가 모두 가능하다.
#ListQueue 클래스는 한쪽(뒤에서 넣고 앞에서 뺌)만 지원했지만,
#Deque 클래스에서는 다음과 같은 메서드를 새로 만들었다:

#- `enqueueFront(x)` : 앞에서 삽입 (리스트의 맨 앞에 넣기)
#- `dequeueRear()` : 뒤에서 삭제 (리스트의 맨 뒤에서 pop)
#- 기존의 `enqueue()` → `enqueueRear()`,
#  `dequeue()` → `dequeueFront()`로 이름만 바꿈

#리스트를 사용했기 때문에 삽입/삭제는 간단하게 처리할 수 있다.
```
