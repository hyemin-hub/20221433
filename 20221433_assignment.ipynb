```python

class Heap:
    def __init__(self, *args):
        if len(args) != 0:
            self.__A = args[0]
        else:
            self.__A = []

    def insert(self, x):
        self.__A.append(x)
        self.__percolateUp(len(self.__A)-1)

    def __percolateUp(self, i: int):
        parent = (i - 1) // 2
        if i > 0 and self.__A[i][0] > self.__A[parent][0]:  # 생일 비교
            self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
            self.__percolateUp(parent)

    def deleteMax(self):
        if not self.isEmpty():
            max = self.__A[0]
            self.__A[0] = self.__A.pop()
            self.__percolateDown(0)
            return max
        else:
            return None

    def __percolateDown(self, i: int):
        child = 2 * i + 1
        right = 2 * i + 2
        if child <= len(self.__A) - 1:
            if right <= len(self.__A) - 1 and self.__A[child][0] < self.__A[right][0]:
                child = right
            if self.__A[i][0] < self.__A[child][0]:
                self.__A[i], self.__A[child] = self.__A[child], self.__A[i]
                self.__percolateDown(child)

    def isEmpty(self) -> bool:
        return len(self.__A) == 0
```

```python

import pandas as pd
from datetime import datetime

# CSV 파일 읽기
df = pd.read_csv("DS_Birthdaydata - 시트1.csv")

# '생년월일8자리(예.20040101)' 열을 datetime으로 변환
df['생일'] = pd.to_datetime(df['생년월일8자리(예.20040101)'], format='%Y%m%d', errors='coerce')

# 날짜 못 바꾼 row 제거
df = df.dropna(subset=['생일'])

# 힙 생성
heap = Heap()

# 힙에 (생일, 이름) 형태로 넣기
for _, row in df.iterrows():
    name = row['이름']
    birthday = row['생일']  # ✅ 이제 이 열을 씀!
    heap.insert((birthday, name))

# 출력
print("🎂 생일이 가장 늦은 친구 TOP 10:")
for i in range(10):
    if not heap.isEmpty():
        bday, name = heap.deleteMax()
        print(f"{i+1}. {name} - {bday.strftime('%Y-%m-%d')}")
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def insertLast(self, data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            newNode.next = newNode
            newNode.prev = newNode
        else:
            tail = self.head.prev
            tail.next = newNode
            newNode.prev = tail
            newNode.next = self.head
            self.head.prev = newNode

    def traverse(self):
        result = []
        if self.head is None:
            return result
        node = self.head
        while True:
            result.append(node.data)
            node = node.next
            if node == self.head:
                break
        return result
```

```python

# 리스트 생성
cdll = CircularDoublyLinkedList()

# 생일 데이터를 리스트에 삽입
for _, row in df.iterrows():
    name = row['이름']
    birthday = row['생일']
    cdll.insertLast((name, birthday))

# 🔄 이미지 기준으로 정확히 반영한 조원 리스트
my_team = ['김예원', '강주영', '정혜민', '윤소정', '홍서연', '이채린', '김소민', '강다원', '전예빈', '이서연']

# 조원만 출력
print("조원 생일:")
node = cdll.head
if node is not None:
    while True:
        name, birthday = node.data
        if name in my_team:
            print(f"{name} - {birthday.strftime('%Y-%m-%d')}")
        node = node.next
        if node == cdll.head:
            break
```

## 5번 - 연습문제 풀기

✅ 01번

풀이:
최대 힙(Max Heap)은 "부모 노드의 값이 자식 노드의 값보다 크거나 같다"는 조건을 만족하는 완전 이진 트리입니다.
루트는 트리의 최상단에 위치하며, 모든 노드의 조상에 해당하므로 전체 힙에서 가장 큰 값을 갖습니다.
따라서 루트보다 더 깊은 곳에 있는 노드가 루트보다 큰 값을 갖는 경우는 절대 존재할 수 없습니다.
이는 힙의 기본 성질에 의해 보장됩니다.

✅ 02번

풀이:
최대 힙은 정렬된 자료구조가 아니라, 부모 ≥ 자식 관계만 보장합니다.
A[0]은 루트 노드이므로 가장 큰 값을 갖지만, A[n-1]은 단지 힙을 배열로 표현했을 때 가장 마지막 위치에 있는 값일 뿐이며,
그 값이 반드시 최소값이라는 보장은 없습니다.
따라서 A[n-1]이 힙에서 가장 작은 값을 가질 가능성은 있지만, 항상 그렇다고 단정할 수는 없습니다.

✅ 03번

풀이:
완전 이진 트리에서 리프 노드(leaf node)는 자식이 없으므로 스며내리기 과정이 필요 없습니다.
힙을 배열로 나타낼 때, 리프 노드는 A[⌊n/2⌋]부터 A[n-1]까지입니다.
따라서 스며내리기가 실제로 수행되는 노드는 루트부터 A[⌊n/2⌋ - 1]까지의 노드이며,
이들의 수는 ⌊n/2⌋개입니다.
즉, 총 n개의 노드 중 절반 정도만 스며내리기를 수행하게 됩니다.

✅ 04번

풀이:
스며내리기(percolateDown)의 시간 복잡도는 노드가 이동할 수 있는 최대 거리, 즉 트리의 높이와 관련이 있습니다.

트리의 높이는 O(log n)이므로,

최악의 경우: 루트에서 리프까지 내려가야 하므로 O(log n)

최선의 경우: 자식보다 큰 경우 한 번 비교 후 종료되므로 O(1)
따라서 전체 시간 복잡도를 Θ(log n) 으로 표현할 수 있습니다.

✅ 05번

풀이:
힙에서 일반적으로 지원하는 삭제 연산은 **루트 삭제(deleteMax 또는 deleteMin)**입니다.
루트를 삭제한 후 마지막 원소를 루트로 이동시키고, 스며내리기를 통해 힙 조건을 복원합니다.
하지만 힙 내에서 임의의 원소(예: 맵의 마지막 원소)를 삭제하려면, 해당 원소의 위치를 찾아야 하며,
삭제 이후에도 힙 조건을 복원하기 위한 추가 작업이 필요합니다.
이 과정은 시간 복잡도 측면에서 비효율적이며, 힙의 원래 용도(우선순위 기반 접근)에도 맞지 않습니다.
결론적으로, 루트 이외의 원소 삭제는 간단하지 않습니다.

✅ 06번

풀이:
buildHeap()은 하위 노드부터 위로 올라오며 스며내리기를 수행하는 방식입니다.
이 방식은 이미 자식 노드가 힙 조건을 만족한 상태에서 부모 노드를 조정하기 때문에 불필요한 작업을 줄일 수 있습니다.
반대로 위에서 아래로 스며내릴 경우, 한 번 조정된 자식 노드가 이후에 다시 스며내려질 수 있어 작업 중복이 발생합니다.
따라서 시간 효율 측면에서 위에서 아래로 스며내리는 방식은 비효율적이며, 교재에서 제시한 방법(아래→위 순서)이
O(n) 시간 내에 힙을 만드는 데 최적화되어 있습니다.

✅ 07번

풀이:
새로운 원소가 힙에 삽입되면, 우선 배열의 마지막 위치(리프의 다음 위치)에 추가됩니다.
이때 힙의 조건(부모 ≥ 자식)이 깨질 수 있으므로, 부모 노드와 비교하면서 위로 이동하는 스며올리기(percolateUp) 과정이 수행됩니다.
각 비교마다 부모와 자리를 교환하며 올라가게 되며,
최악의 경우 루트까지 올라갈 수 있으므로 전체 시간 복잡도는 **O(log n)**입니다.
이 과정을 통해 힙 특성을 유지하게 됩니다.

```python

class MinHeap:
    def __init__(self):
        self.A = []

    def insert(self, x):
        self.A.append(x)
        self._percolate_up(len(self.A) - 1)

    def _percolate_up(self, i):
        parent = (i - 1) // 2
        if i > 0 and self.A[i] < self.A[parent]:
            self.A[i], self.A[parent] = self.A[parent], self.A[i]
            self._percolate_up(parent)

    def delete_min(self):
        if not self.is_empty():
            min_val = self.A[0]
            self.A[0] = self.A.pop()
            self._percolate_down(0)
            return min_val
        return None

    def _percolate_down(self, i):
        child = 2 * i + 1
        right = 2 * i + 2
        if child < len(self.A):
            if right < len(self.A) and self.A[child] > self.A[right]:
                child = right
            if self.A[i] > self.A[child]:
                self.A[i], self.A[child] = self.A[child], self.A[i]
                self._percolate_down(child)

    def min(self):
        return self.A[0] if not self.is_empty() else None

    def size(self):
        return len(self.A)

    def is_empty(self):
        return len(self.A) == 0
```

```python

class KthLargest:
    def __init__(self, k, nums):
        self.k = k
        self.minheap = MinHeap()
        for num in nums:
            self.add(num)

    def add(self, val):
        if self.minheap.size() < self.k:
            self.minheap.insert(val)
        elif val > self.minheap.min():
            self.minheap.delete_min()
            self.minheap.insert(val)
        return self.minheap.min()

#사용예시
# 예시 1
kthLargest = KthLargest(3, [4, 5, 8, 2])
print(kthLargest.add(3))  # 4
print(kthLargest.add(5))  # 5
print(kthLargest.add(10)) # 5
print(kthLargest.add(9))  # 8
print(kthLargest.add(4))  # 8

# 예시 2
kthLargest2 = KthLargest(4, [7, 7, 7, 7, 8, 3])
print(kthLargest2.add(2))   # 7
print(kthLargest2.add(10))  # 7
print(kthLargest2.add(9))   # 7
print(kthLargest2.add(9))   # 8
```
